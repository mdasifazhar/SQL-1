###### Task: Create a sql db from adult dataset and name it sqladb

```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
adult_data =pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',index_col=None,sep='\s*,\s*',
                           header=0, encoding='ascii', engine='python')
adult_data.head(5)
```

#### Rename column

```python
adult_data.columns=["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
adult_data.head(5)
```

#### Create a sql db from adult dataset and name it sqladb

###### 1. Select 10 records from the adult sqladb

```python
con = sqlite3.connect("sqladb")
cur = con.cursor()
```

```python
adult_data.to_sql("sqladb", con, if_exists="replace", index=True)
#cur.execute('drop table sqladb')
```

```python
record=pd.read_sql_query('SELECT * FROM sqladb limit 10',con)
record
```

###### 2. Show  the average hours per week of all men who are working in private sector

```python
#pd.read_sql_query("update sqladb set workclass='NA'where workclass=?",('?',),con)
AvgHrsPerWeek=pd.read_sql_query("SELECT avg(hours_per_week) FROM sqladb WHERE workclass='Private' and sex='Male'",con)
#workclass='Private'
#t=(workclass,)
#cur.execute("SELECT * FROM sqladb WHERE workclass=?",t)
#AvgHrsPerWeek=pd.DataFrame(cur.fetchall())
AvgHrsPerWeek
```

###### 3. Show the frequency table for education, occupation and relationship, separately

```python
education=pd.read_sql_query('SELECT education, COUNT(*) FROM sqladb GROUP BY education',con)
#edu_df= pd.DataFrame(cur.fetchall())
#edu_df.rename(columns={0:'Education',1:'Frequency'},inplace=True)
education
```

```python
occupation=pd.read_sql_query('SELECT occupation, COUNT(*) FROM sqladb GROUP BY occupation',con)
#edu_df= pd.DataFrame(cur.fetchall())
#edu_df.rename(columns={0:'Education',1:'Frequency'},inplace=True)
occupation
```

```python
relationship=pd.read_sql_query('SELECT relationship, COUNT(*) FROM sqladb GROUP BY relationship',con)
#rel_df= pd.DataFrame(cur.fetchall())
#rel_df.rename(columns={0:'Relationship',1:'Frequency'},inplace=True)
relationship
```

###### 4. Are there any people who are married, working in private sector and having a masters degree

```python
#people=pd.read_sql_query("SELECT distinct(marital_status) FROM sqladb ",con)

people=pd.read_sql_query("SELECT COUNT(*) as Married FROM sqladb WHERE  marital_status!='Never-married' and workclass='Private' and education='Masters'",con)
#rel_df= pd.DataFrame(cur.fetchall())
#rel_df.rename(columns={0:'Relationship',1:'Frequency'},inplace=True)
people
```

###### 5. What is the average, minimum and maximum age group for people working in different sectors

```python
people_diff_sec=pd.read_sql_query("SELECT workclass, avg(age), min(age), max(age) FROM sqladb GROUP BY workclass",con)
people_diff_sec
```

###### 6. Calculate age distribution by country

```python
dis=pd.read_sql_query("SELECT native_country, min(age) As Min,max(age) AS Max,avg(age) AS Avg FROM sqladb GROUP BY native_country",con)
dis
```

###### 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
NetCapitalGain=pd.read_sql_query("SELECT *,capital_gain-capital_loss AS Net_Capital_Gain FROM sqladb order by Net_Capital_Gain desc",con)
NetCapitalGain.head(5)
```
